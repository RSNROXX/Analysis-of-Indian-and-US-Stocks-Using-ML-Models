```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# from sklearn.model_selection import cross_val_score, train_test_split, GridSearchCV
# from sklearn.feature_selection import RFECV, SelectFromModel, SelectKBest
# from sklearn.preprocessing import StandardScaler
# from sklearn import metrics

import yfinance as yf
import seaborn as sns
sns.set_style('whitegrid')
plt.style.use("fivethirtyeight")
%matplotlib inline

df = yf.download('^NSEI', start='2019-01-01', end='2019-06-14')
df.tail()
```

```python
# x=[2019-05-29,2019-05-31,2019-06-03,2019-06-04,2019-06-05,2019-06-06,2019-06-07,2019-06-10,2019-06-11,2019-06-12,2019-06-13]
# y = [7520.84,7400.65,7395.71,7512.63,7460.72,7609.19,7654.80,7524.75,7963.27,7682.50,7655.81]
plt.figure(figsize=(12,7))
plt.title('Nifty')
plt.xlabel('Days')
plt.ylabel('Opening Price USD ($)')
plt.plot(df['Close'], label='Nifty')
plt.legend()
plt.show()
```

```python
df2 = df['Close']
df2 = pd.DataFrame(df2)
```

```python
future_days = 11
df2['Prediction'] = df2['Close'].shift(-future_days)
```

```python
df2.tail()
```

```python
X = np.array(df2.drop(['Prediction'], 1))[:-future_days]
print(X)
```

```python
predictions = [11494.74,11520.37,11532.68,11548.43,11562.54,11452.81,11648.92,11507.68,11874.89,11895.98,11905.28]
valid = df2[X.shape[0]:]
valid['Predictions'] = predictions
```

```python
plt.figure(figsize=(12,7))
plt.title("Random forest prediction")
plt.xlabel('Days')
plt.ylabel('Close Price USD ($)')
plt.plot(df2['Close'],label='Nasdaq')
plt.plot(valid['Predictions'], label='RF',color='green')
# plt.legend(["Original", "Valid", 'Predicted'])
plt.legend()
plt.show()
```
